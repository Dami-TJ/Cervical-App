```python

import warnings
warnings.filterwarnings("ignore")
import os
# os.environ['CUDA_VISIBLE_DEVICES'] = '-1'  # Comment out to enable GPU if desired

import pandas as pd
import numpy as np
import tensorflow as tf
from math import ceil
import matplotlib.pyplot as plt
import hashlib
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.applications.densenet import preprocess_input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Function to load data
def load_data(dataset_dir):
    file_paths, labels = [], []
    has_subfolders = False
    for root, dirs, files in os.walk(dataset_dir):
        class_name = os.path.basename(root)
        if class_name == ".ipynb_checkpoints":
            continue
        if root != dataset_dir:
            has_subfolders = True
        image_files = [file for file in files if file.endswith((".jpg", ".png", ".jpeg"))]
        for file in image_files:
            try:
                file_path = os.path.join(root, file)
                with open(file_path, 'rb') as f:
                    _ = f.read()
                file_paths.append(file_path)
                labels.append(class_name if has_subfolders else None)
            except Exception as e:
                print(f"Error reading file {file_path}: {e}")
    df = pd.DataFrame({"file_path": file_paths, "label": labels})
    print(f"Loaded {len(df)} images from {dataset_dir}")
    print("Labels found:", df["label"].unique())
    if not has_subfolders:
        print(f"Warning: No class subfolders found in {dataset_dir}. 'label' column may be invalid.")
    return df

# Alternative load_data for flat directory with CSV labels (uncomment if needed)
"""
def load_data(dataset_dir, csv_path=None):
    file_paths, labels = [], []
    if csv_path:
        try:
            label_df = pd.read_csv(csv_path)
            print(f"Loaded CSV with {len(label_df)} entries from {csv_path}")
            for _, row in label_df.iterrows():
                file_path = os.path.join(dataset_dir, row['filename'])
                if os.path.exists(file_path) and file_path.endswith((".jpg", ".png", ".jpeg")):
                    try:
                        with open(file_path, 'rb') as f:
                            _ = f.read()
                        file_paths.append(file_path)
                        labels.append(str(row['label']))
                    except Exception as e:
                        print(f"Error reading file {file_path}: {e}")
        except Exception as e:
            print(f"Error reading CSV {csv_path}: {e}")
    else:
        has_subfolders = False
        for root, dirs, files in os.walk(dataset_dir):
            class_name = os.path.basename(root)
            if class_name == ".ipynb_checkpoints":
                continue
            if root != dataset_dir:
                has_subfolders = True
            image_files = [file for file in files if file.endswith((".jpg", ".png", ".jpeg"))]
            for file in image_files:
                try:
                    file_path = os.path.join(root, file)
                    with open(file_path, 'rb') as f:
                        _ = f.read()
                    file_paths.append(file_path)
                    labels.append(class_name if has_subfolders else None)
                except Exception as e:
                    print(f"Error reading file {file_path}: {e}")
    df = pd.DataFrame({"file_path": file_paths, "label": labels})
    print(f"Loaded {len(df)} images from {dataset_dir}")
    print("Labels found:", df["label"].unique())
    if not csv_path and not has_subfolders:
        print(f"Warning: No class subfolders found in {dataset_dir}. 'label' column may be invalid.")
    return df
"""

# Load original dataset
dataset_dir = "./data"
df = load_data(dataset_dir)
if df.empty or 'label' not in df.columns or df["label"].isna().all():
    raise ValueError("Original dataset is empty or missing valid 'label' column")
df['hash'] = df['file_path'].apply(lambda x: hashlib.md5(open(x, 'rb').read()).hexdigest())
print("Number of duplicates in original data:", df['hash'].duplicated().sum())
df = df.drop_duplicates(subset='hash').drop(columns='hash')
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Balance original dataset
class_counts = df["label"].value_counts()
print("Class Distribution Before Balancing:\n", class_counts)
min_samples = class_counts.min()
balanced_df = df.groupby("label").apply(lambda x: x.sample(min_samples, random_state=42)).reset_index(drop=True)
print("Class Distribution After Balancing:\n", balanced_df["label"].value_counts())

# Load new test dataset
new_test_dir = "./balanced2"
new_test_df = load_data(new_test_dir)
if new_test_df.empty or 'label' not in new_test_df.columns or new_test_df["label"].isna().all():
    raise ValueError("New test dataset is empty or missing valid 'label' column")
new_test_df['hash'] = new_test_df['file_path'].apply(lambda x: hashlib.md5(open(x, 'rb').read()).hexdigest())
print("Number of duplicates in new test data:", new_test_df['hash'].duplicated().sum())
new_test_df = new_test_df.drop_duplicates(subset='hash').drop(columns='hash')
print("New Test Data Class Distribution:\n", new_test_df["label"].value_counts())

# Load original test dataset
test_dir = "./test_data"
test_df = load_data(test_dir)
use_test_data = True
if test_df.empty or 'label' not in test_df.columns or test_df["label"].isna().all():
    print("Warning: Test dataset is empty or missing valid 'label' column. Skipping original test data evaluation.")
    use_test_data = False
else:
    test_df['hash'] = test_df['file_path'].apply(lambda x: hashlib.md5(open(x, 'rb').read()).hexdigest())
    print("Number of duplicates in test data:", test_df['hash'].duplicated().sum())
    test_df = test_df.drop_duplicates(subset='hash').drop(columns='hash')
    print("Test Data Class Distribution:\n", test_df["label"].value_counts())

# Parameters
IMG_SIZE = (224, 224)
CHANNELS = 3
IMG_SHAPE = (IMG_SIZE[0], IMG_SIZE[1], CHANNELS)
BATCH_SIZE = 64
TEST_BATCH_SIZE = BATCH_SIZE
EPOCHS = 10
K_FOLDS = 5
test_length = len(test_df) if use_test_data else 0

# Test data generator (only if test data is valid)
test_gen = None
if use_test_data:
    test_gen_aug = ImageDataGenerator(preprocessing_function=preprocess_input)
    test_gen = test_gen_aug.flow_from_dataframe(
        test_df, x_col="file_path", y_col="label",
        target_size=IMG_SIZE, batch_size=TEST_BATCH_SIZE, class_mode="categorical",
        shuffle=False, seed=42
    )

# K-Fold setup on original data
skf = StratifiedKFold(n_splits=K_FOLDS, shuffle=True, random_state=42)
all_train_acc, all_val_acc, all_train_loss, all_val_loss = [], [], [], []
fold_no = 1

# Create checkpoint directory
checkpoint_dir = "./checkpoints"
os.makedirs(checkpoint_dir, exist_ok=True)

for train_index, val_index in skf.split(balanced_df["file_path"], balanced_df["label"]):
    print(f"\nTraining Fold {fold_no}/{K_FOLDS}")
    train_df, val_df = balanced_df.iloc[train_index], balanced_df.iloc[val_index]
    print("Training class distribution:\n", train_df['label'].value_counts())
    print("Validation class distribution:\n", val_df['label'].value_counts())
    
    # Data generators
    train_gen_aug = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.4,
        height_shift_range=0.4,
        shear_range=0.4,
        zoom_range=0.4,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest',
        preprocessing_function=preprocess_input
    )
    val_gen_aug = ImageDataGenerator(preprocessing_function=preprocess_input)
    
    train_gen = train_gen_aug.flow_from_dataframe(
        train_df, x_col="file_path", y_col="label",
        target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode="categorical",
        shuffle=True, seed=42
    )
    val_gen = val_gen_aug.flow_from_dataframe(
        val_df, x_col="file_path", y_col="label",
        target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode="categorical",
        shuffle=False, seed=42
    )
    
    # Check class consistency with test data (if available)
    if use_test_data and set(test_gen.class_indices.keys()) != set(train_gen.class_indices.keys()):
        raise ValueError("Class labels in original test data do not match training data classes.")
    
    # Model
    base_model = DenseNet121(input_shape=IMG_SHAPE, weights='imagenet', include_top=False)
    for layer in base_model.layers:
        layer.trainable = True
    
    model = Sequential([
        base_model,
        Flatten(),
        BatchNormalization(),
        Dense(64, activation="relu", kernel_regularizer=l2(0.01)),
        Dropout(0.3),
        BatchNormalization(),
        Dense(32, activation="relu", kernel_regularizer=l2(0.01)),
        Dropout(0.3),
        Dense(len(train_gen.class_indices), activation="softmax")
    ])
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    
    # Callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    checkpoint = ModelCheckpoint(
        os.path.join(checkpoint_dir, f'best_model_fold_{fold_no}.keras'),
        save_best_only=True
    )
    
    # Training
    history = model.fit(
        train_gen, epochs=EPOCHS, validation_data=val_gen,
        callbacks=[early_stop, checkpoint], verbose=1
    )
    
    all_train_acc.append(history.history["accuracy"])
    all_val_acc.append(history.history["val_accuracy"])
    all_train_loss.append(history.history["loss"])
    all_val_loss.append(history.history["val_loss"])

    #orginal data confusion matrix
    y_pred = np.argmax(model.predict(val_gen), axis=1)
    y_true = val_gen.classes
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=val_gen.class_indices.keys())
    plt.figure(figsize=(6, 5))
    disp.plot(cmap="Blues", xticks_rotation=45)
    plt.title(f"Confusion Matrix - Fold {fold_no}")
    plt.show()
    plt.figure(figsize=(12, 5))

    # Plot cumulative training and validation metrics up to this fold
    plt.figure(figsize=(12, 5))
    
    # Training vs Validation Accuracy
    plt.subplot(1, 2, 1)
    for i in range(fold_no):
        epochs_range = range(1, len(all_train_acc[i]) + 1)
        plt.plot(epochs_range, all_train_acc[i], label=f"Train Fold {i+1}")
        plt.plot(epochs_range, all_val_acc[i], label=f"Val Fold {i+1}", linestyle="dashed")
    plt.title(f"Training & Validation Accuracy (Folds 1-{fold_no})")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    
    # Training vs Validation Loss
    plt.subplot(1, 2, 2)
    for i in range(fold_no):
        epochs_range = range(1, len(all_train_loss[i]) + 1)
        plt.plot(epochs_range, all_train_loss[i], label=f"Train Fold {i+1}")
        plt.plot(epochs_range, all_val_loss[i], label=f"Val Fold {i+1}", linestyle="dashed")
    plt.title(f"Training & Validation Loss (Folds 1-{fold_no})")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Test on new dataset
    print(f"\nTesting Fold {fold_no} on New Data...")
    new_test_gen = val_gen_aug.flow_from_dataframe(
        new_test_df, x_col="file_path", y_col="label",
        target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode="categorical",
        shuffle=False, seed=42
    )
    
    if set(new_test_gen.class_indices.keys()) != set(train_gen.class_indices.keys()):
        raise ValueError("Class labels in new test data do not match training data classes.")
    
    y_pred_new = np.argmax(model.predict(new_test_gen), axis=1)
    y_true_new = new_test_gen.classes
    cm_new = confusion_matrix(y_true_new, y_pred_new)
    disp_new = ConfusionMatrixDisplay(cm_new, display_labels=list(new_test_gen.class_indices.keys()))
    
    plt.figure(figsize=(6, 5))
    disp_new.plot(cmap="Blues", xticks_rotation=45)
    plt.title(f"Confusion Matrix - Fold {fold_no} (New Test Data)")
    plt.show()
    
    print(f"\nClassification Report - Fold {fold_no} (New Test Data):")
    print(classification_report(y_true_new, y_pred_new, target_names=list(new_test_gen.class_indices.keys())))
    
    # Test on original test data (if available)
    if use_test_data:
        print(f"\nTesting Fold {fold_no} on Original Test Data...")
        test_steps = ceil(test_length / TEST_BATCH_SIZE)
        test_score = model.evaluate(test_gen, steps=test_steps)
        print(f"Original Test Loss: {test_score[0]:.6f}")
        print(f"Original Test Accuracy: {round(test_score[1]*100)}%")
        
        Yhat = np.argmax(model.predict(test_gen, steps=test_steps), axis=1)
        Ytrue = test_gen.classes
        cm = confusion_matrix(Ytrue, Yhat)
        disp = ConfusionMatrixDisplay(cm, display_labels=list(test_gen.class_indices.keys()))
        
        fig, ax = plt.subplots(figsize=(10, 8))
        disp.plot(cmap="Blues", ax=ax)
        ax.grid(False)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
        plt.title(f"Confusion Matrix - Fold {fold_no} (Original Test Data)")
        plt.show()
        
        print("\nClassification Report - Original Test Data:")
        print(classification_report(Ytrue, Yhat, target_names=list(test_gen.class_indices.keys())))
    
    fold_no += 1

# Summary statistics
mean_val_acc = np.mean([acc[-1] for acc in all_val_acc])
std_val_acc = np.std([acc[-1] for acc in all_val_acc])
print(f"\nMean Validation Accuracy (Original Data): {mean_val_acc:.4f} ± {std_val_acc:.4f}")
```
